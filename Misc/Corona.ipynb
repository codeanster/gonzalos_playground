```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 28})
```

```python
#data is structured by date, let's look at data from 01-22-2019
datadir = '/home/gonzalo/gonzalos_playground/COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/'

#sample date
date = '01-22-2020.csv'

#reads data in pandas dataframe
data = pd.read_csv(datadir + date)
location = np.nan_to_num(data['Country/Region']) #locations of reports
deaths = (data['Deaths']) #total deaths
infected = data['Confirmed'] #infected


r_ind = []
for i in range(len(location)):
    if deaths[i] >5:
        print(deaths[i],i)
        r_ind.append(i)
    if location[i] == 'Mainland China':
        location[i] = 'China'


#plotting parameters
plt.figure(figsize=(20,12))
plt.title(date[:-4])
plt.bar(location[r_ind],deaths[r_ind],width=0.2,label='Deaths')
plt.bar(location[r_ind],infected[r_ind],width=-0.2,align='edge',color='red',label='Infected') 
plt.legend(prop={'size': 40})
plt.savefig('corona{}.png'.format(date))
```

```python
#Let's create a system to loop through all the dates in the month

def corona_time_series(month,days):
    """This function will take a given month, expressed as an integer from 1 to 12, and the number of days
    starting from 1, and make a plot for each day of countries showing deaths from the corona virus greater
    than five."""

    days = np.arange(1,days)
    print(len(days))
    year = 2020
    for i in range(len(days)):
        date = '{}-{}-{}.csv'.format(month,days[i],year)
        if i < 9:

            #So apparently it's difficult to add a 0 into a string that's not leading or trailing. So im just going to break 
            #them up
            date = date.rjust(1 + len(date),'0')
            str1 = date[0:3]
            str2 = date[3:13]

            #add a leading zero to the second term 
            str2 = str2.rjust(1+len(str2),'0')
            date = str1 + str2
        else:
            date = date.rjust(1 + len(date),'0')
            
        data = pd.read_csv(datadir + date)
        print(date)
        location = np.nan_to_num(data['Country/Region']) #locations of reports
        deaths = (data['Deaths']) #total deaths
        infected = data['Confirmed'] #infected

        r_ind = []
        for i in range(len(location)):
            if deaths[i] >5:
                r_ind.append(i)
            if location[i] == 'Mainland China':
                location[i] = 'China'


        plt.figure(figsize=(20,12))
        plt.title(date[:-4])
        plt.bar(location[r_ind],deaths[r_ind],width=0.2,label='Deaths')
        plt.bar(location[r_ind],infected[r_ind],width=-0.2,align='edge',color='red',label='Infected')

        plt.legend(prop={'size': 40})
        plt.savefig('corona{}.png'.format(date))
        plt.close()
        plt.plot(np.arange(len(days)),)
```
