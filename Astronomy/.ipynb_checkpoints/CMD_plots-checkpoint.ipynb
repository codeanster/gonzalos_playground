This notebook will attempt to repicate the graph from 
Welcome to the twilight Zone: The Mid-infrared Properties of Post-starburst Galaxies

The SPOGS survey was nice enough to release their data publicly here: http://www.spogs.org/spogtable.txt

```python
#packages needed to get this notebook to run

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astroquery.vizier import Vizier
from marvin.tools.query import Query
from scipy.optimize import curve_fit
from pylab import *


plt.rcParams.update({'font.size': 20})
```

```python
#useful functions

def gauss(mx,mu,sigma,A):
    return A*exp(-(mx-mu)**2/2/sigma**2)

def bimodal(mx,mu1,sigma1,A1,mu2,sigma2,A2):
    return gauss(mx,mu1,sigma1,A1)+gauss(mx,mu2,sigma2,A2)
```

```python
Vizier.ROW_LIMIT = -1 # we have to put this so we can get all the rows
table = Vizier.get_catalogs(catalog='J/ApJS/224/38/table2')[0] #Query with Vizer with catalog from link
```

```python
color = table['gmag'] - table['rmag']
mass = table['logM_']

#galaxies representated in the Green valley (maybe)
plt.figure(figsize=(8,10))
plt.plot(mass,color,'go')
plt.ylim(0.14,1.4)
plt.xlabel(r'$Log(M_\star / M_\odot)$')
plt.ylabel('g-r')
plt.title('Spogs Galaxy Sample')
```

```python
#now we have the table. I want to know select galaxies from the MaNGA survey
sf = 'nsa.z > 0'
q = Query(search_filter=sf, limit=10000,return_params=['nsa.iauname','nsa.sersic_logmass','nsa.elpetro_mag_g_r'])


r = q.run()
manga_data = r.toTable() #sample of 4679 galaxies
manga_mass = manga_data['nsa.sersic_logmass']
manga_color = manga_data['nsa.elpetro_mag_g_r']
```

```python
#galaxies in MaNGA survey

plt.figure(figsize=(8,10))
plt.plot(manga_mass,manga_color,'o') 
plt.ylim(0.14,1.4)
plt.xlabel(r'$Log(M_\star / M_\odot)$')
plt.ylabel('g-r')
plt.title('MaNGA Galaxy Sample')
```

```python
#create a 1 x 3 plot for MaNGA, SPOGS, and each overlayed

fig, axes = plt.subplots(figsize=(16,14),nrows=1, ncols=3, sharex=True,sharey=True)

axes[0].plot(manga_mass,manga_color,'o') 
axes[0].set_ylim(0.14,1.4)
axes[0].set_xlabel(r'$Log(M_\star / M_\odot)$')
axes[0].set_ylabel('g-r')
axes[0].set_title('MaNGA Galaxy Sample')

axes[1].plot(mass,color,'go') 
axes[1].set_ylim(0.14,1.4)
axes[1].set_xlabel(r'$Log(M_\star / M_\odot)$')
axes[1].set_ylabel('g-r')
axes[1].set_title('Spogs Galaxy Sample')


axes[2].plot(manga_mass,manga_color,'o') 
axes[2].set_ylim(0.14,1.4)
axes[2].set_xlabel(r'$Log(M_\star / M_\odot)$')
axes[2].set_ylabel('g-r')
axes[2].set_title('Overlay')
axes[2].plot(mass,color,'go') 

plt.savefig('CMD_plots')
```

```python
#Plot a histogram
y,x,_=hist(manga_color,150,alpha=1,label='data')
mx=(x[1:]+x[:-1])/2 # for len(x)==len(y)
expected=(0.4,.1,110,0.7,.1,300) #guess parameters for each gaussian : mean, std, amp

#solve for parameters using scipy curvy fit, guess values and the data
params,cov=curve_fit(bimodal,mx,y,expected)
sigma=sqrt(diag(cov))
plot(mx,bimodal(mx,*params),color='red',lw=3,label='model') #plotting the double gaussian
legend()
print(params,'\n',sigma)

#plot parameters
plt.title('MaNGA Survey Sample')
plt.xlabel('g-r')
plt.ylabel('Number of Galaxies')
plt.xlim(0,1.4)
plt.tight_layout()
plt.savefig('MaNGA_hist.png')
```
